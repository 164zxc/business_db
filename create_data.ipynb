```python
updated_columns = [
    'STUDENT_NO', 'SEXUAL', 'AGE', 'SUBJECT_NAME', 'EXAM_TYPE',
    'COURSE_TYPE_SUB', 'TEACHER', 'ORDER_PRICE', 'CREATE_DATE',
    'PROGRAM_TYPE', 'LOCATION', 'STATUS',
    'IS_OLD_CUSTOMER(IF PREVIOUS ORDER EXISTS)', 'INVOICE_NO',
    'STUDENT_NO_DETAIL_ITEM', 'ORDER_TOTAL', 'DISCOUNT_AMOUNT',
    'CLERK_NO'
]

# Maintain a mapping for CLERK_NO and its corresponding LOCATION
clerk_location_mapping = {}
student_registry = {}
def get_or_generate_student_details(student_no):
    if student_no not in student_registry:
        student_registry[student_no] = {
            "STUDENT_NO": student_no,
            "SEXUAL": random.choice(['Male', 'Female']),
            "AGE": random.randint(15, 40)
        }
    return student_registry[student_no]
# Helper function to get or assign a LOCATION for a given CLERK_NO
def get_or_assign_location(clerk_no):
    if clerk_no not in clerk_location_mapping:
        clerk_location_mapping[clerk_no] = random.choice(['Region A', 'Region B', 'Region C', 'Region D', 'Region E', 'Region F', 'Region G', 'Region H'])
    return clerk_location_mapping[clerk_no]

# Generate student orders with consistent LOCATION for each CLERK_NO
def generate_student_orders_with_clerk_location(student_details):
    student_data = []
    num_products = random.randint(1, 3)  # Each student buys 1-3 products
    clerk_no = f"CL-{random.randint(100, 200)}"
    location = get_or_assign_location(clerk_no)  # Get or assign LOCATION for this CLERK_NO
    create_date = (datetime.now() - timedelta(days=random.randint(1,730))).strftime('%Y-%m-%d')
    is_old_customer = random.choice(['New', 'Old'])

    for _ in range(num_products):
        invoice_no = f"INV-{random.randint(100000, 999999)}"
        student_no_detail_item = invoice_no  # Same as INVOICE_NO
        program_type = random.choice(['Exam01', 'Exam02','Exam03','Exam04','Exam05','Exam06'])
        status = random.choice(['本班'])
        
        num_courses = random.randint(1, 5)  # 1-5 courses per product
        order_total = 0  # Initialize ORDER_TOTAL for the product
        discount_amount = random.randint(1000, 10000)  # Discount for the product

        for _ in range(num_courses):
            subject_name = random.choice(['Math', 'English', 'Science', 'History', 'Art'])
            course_type_sub = random.choice(['面授', '線上', '講座', '學堂', '題庫', '雲端', '衝刺班', '短期'])
            teacher = random.choice(['Alice', 'Angela', 'Bella', 'Cindy', 'Charlotte', 'Cynthia', 'Daisy'])
            order_price = random.randint(10000, 15000)
            order_total += order_price

            # Append row for each course
            student_data.append([
                student_details["STUDENT_NO"],            # STUDENT_NO
                student_details["SEXUAL"],               # SEXUAL
                student_details["AGE"],                  # AGE
                subject_name,                             # SUBJECT_NAME
                random.choice(['Midterm', 'Final']),      # EXAM_TYPE
                course_type_sub,                          # COURSE_TYPE_SUB
                teacher,                                  # TEACHER
                order_price,                              # ORDER_PRICE
                create_date,                              # CREATE_DATE
                program_type,                             # PROGRAM_TYPE
                location,                                 # LOCATION
                status,                                   # STATUS
                is_old_customer,                          # IS_OLD_CUSTOMER
                invoice_no,                               # INVOICE_NO
                student_no_detail_item,                   # STUDENT_NO_DETAIL_ITEM
                '',                                       # ORDER_TOTAL (filled later for consistency)
                discount_amount,                          # DISCOUNT_AMOUNT
                clerk_no,                                 # CLERK_NO
           
            ])

        # Update ORDER_TOTAL for all rows of this product
        for row in student_data[-num_courses:]:
            row[15] = order_total - discount_amount  # ORDER_TOTAL = SUM(ORDER_PRICE) - DISCOUNT_AMOUNT

    return student_data

# Generate data for multiple students with consistent CLERK_NO and LOCATION mapping
data = []
student_list = [f"SN-{random.randint(10000, 99999)}" for _ in range(10000)]  # 5 students
for student_no in student_list:
    student_details = get_or_generate_student_details(student_no)
    data.extend(generate_student_orders_with_clerk_location(student_details))

# Create a DataFrame
df = pd.DataFrame(data, columns=updated_columns)
```
